```python
import json
import pandas as pd
```

```python
data = pd.read_csv("ParaMAWPS - ParaMAWPS2.csv")
data = data[['id', 'original_text', 'segmented_text', 'equation', 'ans']]
data
```

```python
data['segmented_text'] = data['original_text']
data
```

```python
def dfToList(df):
    ret = []
    for row in df.iterrows():
        rowDict = row[1].to_dict()
        ret.append(rowDict)

    return ret
```

```python
jsonString = json.dumps(dfToList(data[:13023]))
jsonFile = open("trainset.json", "w")
jsonFile.write(jsonString)
jsonFile.close()
```

```python
jsonString = json.dumps(dfToList(data[13023:14650]))
jsonFile = open("validset.json", "w")
jsonFile.write(jsonString)
jsonFile.close()
```

```python
jsonString = json.dumps(dfToList(data[14650:]))
jsonFile = open("testset.json", "w")
jsonFile.write(jsonString)
jsonFile.close()
```
