````python
import re
import csv

allnums = set()

f = open('file1-3.csv', 'r')
data = csv.reader(f)

for row in data:
    text = row[1]
    for word in text.split():
        if any(c.isdigit() for c in word):
            allnums.add(word)

allnums = list(allnums)
````

````python
newnums = []
remnums = []
for s in allnums:
    if not re.match(r'\$[0-9]+\.[0-9]+$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]+\.[0-9]+\.$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]+\.[0-9]+,$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]{1,3}(,[0-9]{3})+$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]{1,3}(,[0-9]{3})+\.$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]{1,3}(,[0-9]{3})+,$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]+$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]+\.$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]+,$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]+\?$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]+\?$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\$[0-9]+\.[0-9]+\?$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\(\$[0-9]+\.[0-9]+\)[\.,]$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'\([0-9]+\)\??$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'-[0-9]+[\.\?]?$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'[0-9]+[\.,\?]?$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'[0-9]+\.[0-9]+\.?$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'0.25-pound[\.,\?]?$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'[0-9]+-((day)|(year)|(hour)|(mile)|(dollar))$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'[0-9]+/[0-9]+$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'[0-9]{1,3}(,[0-9]{3})+$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'[0-9]+%[\.,\?]?$', s) == None:
        # newnums.append(s)
        continue
    if not re.match(r'[0-9]+\.[0-9]+%?$', s) == None:
        # newnums.append(s)
        continue

    remnums.append(s)

print(f'Matched: {len(newnums)} elements out of {len(allnums)}')
print(newnums)
print(f'Remaining: {len(remnums)} elements')
print(remnums)
````

## Observations
- `$` might appear often before numbers
- number words commonly end with `.`, `,`, or `?`
- float, int, and segmented int can be generalized with groups
- percentage signs may appear at the end of numbers before punctuation
- negative signs may appear right before numbers
- parentheses may appear around the number (so far, only before punctuation)
- all others can be normally converted


### Generalized regex pattern

```python
ptn = r'\(?-?\$?((\d+)|(\d+\.\d+)|(\d{1,3}(,\d{3})*))%?\)?[\.,\?]?$'
```


````python
ptn = r'\(?-?\$?((\d+)|(\d+\.\d+)|(\d{1,3}(,\d{3})*))%?\)?[\.,\?]?$'

remnums = []
newnums = []

for s in allnums:
    if not re.match(ptn, s) == None:
        newnums.append(s)
        continue
    remnums.append(s)

print(f'Matched: {len(newnums)} elements out of {len(allnums)}')
print(newnums)
print(f'Remaining: {len(remnums)} elements')
print(remnums)
````
